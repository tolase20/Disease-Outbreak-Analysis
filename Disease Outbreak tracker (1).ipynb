DISEASE OUTBREAK TRACKER

```python
#Task 1 - Create Variables
disease_name = "cholera"
state = "Lagos"
total_cases =850
new_cases = 45
recovered_patients = 730
deaths = 18
isolation_beds = 95
minimum_safe_bed_capacity = 80
emergency_declared = False
```

Task 2 - Calculate Active Cases

```python
active_cases = total_cases - recovered_patients - deaths
print("Active cases:", active_cases)
```

Task 3 - Calculate Recovery Rate

```python
lagos_recovery_rate = (recovered_patients / total_cases) * 100
print("Recovery Rate:",lagos_recovery_rate)
```

Task 4 -Update Total Cases

```python
updated_total_cases= total_cases + 15
print("Updated total cases:", updated_total_cases)
```

Task 5 - Check Bed Capacity

```python
sufficient_capacity=isolation_beds>minimum_safe_bed_capacity
print ("Hospital capacity sufficient:", isolation_beds>minimum_safe_bed_capacity)
```

Task 6 - Determine if the Outbreak is Severe

```python
severe_outbreak = active_cases>100 or deaths>20
print("Severe Outbreak:", active_cases>100 or deaths>20)
```

Task 7 - Determine if emergency response is Needed

```python
emergency_response = severe_outbreak and not emergency_declared
print("Emergency Response Needed:", severe_outbreak and not emergency_declared)
```

Task 8 - Determine if the situation is stable

```python
Situation_stable = lagos_recovery_rate> 80 and sufficient_capacity
print("Situation stable:", lagos_recovery_rate>80 and sufficient_capacity)
```

Task 9 - Generate the Outbreak Report

```python
print("="*50)
print("                DISEASE OUTBREAK REPORT")
print("="*50)
print("Disease:",disease_name)
print("State:", state)
print("Total Reported cases:",total_cases)
print("New cases:", new_cases)
print("Recovered Patients:", recovered_patients)
print("Deaths:", deaths)
print("Available Beds:",isolation_beds)
print("Hospital Capacity Sufficient:", minimum_safe_bed_capacity)
print("Emergency Declared:", emergency_declared)
print("Active cases:", active_cases)
print("Recovery Rate:",recovery_rate)
print("Updated total cases:", updated_total_cases)
print ("Hospital capacity sufficient:", isolation_beds>minimum_safe_bed_capacity)
print("Severe Outbreak:", active_cases>100 or deaths>20)
print("Emergency Response Needed:", severe_outbreak and not emergency_declared)
print("Situation stable:", recovery_rate>80 and sufficient_capacity)
print("="*50)
```
